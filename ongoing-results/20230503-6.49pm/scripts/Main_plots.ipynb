```python
# this file plots all saved results and combine in one plot
import matplotlib.pyplot as plt
from numpy import loadtxt
import seaborn as sns
```

```python
Average_reward_003 = loadtxt('Average_reward_003.csv', delimiter=',')
Average_reward_altitude = loadtxt('Average_reward_altitude.csv', delimiter=',')
#Average_reward_004 = loadtxt('Average_reward_004.csv', delimiter=',')
Average_reward_fixed_power = loadtxt('Average_reward_fixed_power.csv', delimiter=',')
fig1 = plt.figure()
sns.set_style("whitegrid")
plt.plot(Average_reward_003,linewidth=1)
plt.plot(Average_reward_altitude,linewidth=1)
#plt.plot(Average_reward_004,linewidth=1)
plt.plot(Average_reward_fixed_power,linewidth=1)
plt.xlabel('Number of Episodes')
plt.ylabel('Average Reward')
plt.legend(('NOMA-DQN(altitude=100-140m)','NOMA-DQN(altitude=60-100m)','NOMA-FixedPower'))
```

```python
# reward per episode
reward_per_episode_003 = loadtxt('reward_per_episode_003.csv', delimiter=',')
reward_per_episode_altitude = loadtxt('reward_per_episode_altitude.csv', delimiter=',')
#Average_reward_004 = loadtxt('Average_reward_004.csv', delimiter=',')
reward_per_episode_fixed_power = loadtxt('reward_per_episode_fixed_power.csv', delimiter=',')
fig1 = plt.figure()
sns.set_style("whitegrid")
#plt.plot(reward_per_episode_003,linewidth=1)
#plt.plot(reward_per_episode_altitude,linewidth=1)
#plt.plot(Average_reward_004,linewidth=1)
plt.plot(reward_per_episode_fixed_power,linewidth=1)
plt.xlabel('Number of Episodes')
plt.ylabel('Reward')
plt.legend(('NOMA-DQN(altitude=100-140m)','NOMA-DQN(altitude=60-100m)','NOMA-FixedPower'))
```

```python
losses_per_episode_003=loadtxt('losses_per_episode_003.csv', delimiter=',')
#losses_per_episode_004=loadtxt('losses_per_episode_004.csv', delimiter=',')
#losses_per_episode_fixed_power=loadtxt('losses_per_episode_fixed_power.csv', delimiter=',')
fig2 = plt.figure()
sns.set_style("whitegrid")
#plt.plot(losses_per_episode_004,linewidth=1)
plt.plot(losses_per_episode_003,linewidth=1)
#plt.plot(losses_per_episode_fixed_power,linewidth=1)
plt.xlabel('Number of Episodes')
plt.ylabel('Loss Function')
plt.legend(('Learning Rate=0.004','Learning Rate=0.003','FixedPower'))
```
